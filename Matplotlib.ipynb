# Matplotlib
 <br>
 <br>
Matplotlib 是祖父級的 Python 數據可視化庫。它是由約翰亨特創建的。他創建它是為了嘗試在 Python 中復制 MatLab（另一種編程語言）的繪圖功能。所以如果你碰巧熟悉 matlab，matplotlib 會讓你感覺很自然。

他是優秀的 2D、3D 圖形庫，用於產生科學數據。

**Matplotlib 的一些主要優點是：**

- 對於簡單的繪圖，通常很容易上手
- 支持自定義標籤和文本
- 對圖形中每個元素的出色控制
- 多種格式的高質量輸出
- 一般非常可定制

官方網站: http://matplotlib.org/

# 物件導向 Matplotlib Object Oriented Method
使用更正式的物件導向主要思想是創建圖形對象，然後只調用該對象的方法或屬性。在處理具有多個圖案時，這種方法更好。
## `plt.figure()` : 建立一個空白畫布

```python
fig = plt.figure()
```

#### left : 距離畫布最左邊的長度

#### width : 圖表寬度
#### height : 圖表高度

## 在圖表上建立子圖表

```python
# 新建一個畫布
fig = plt.figure()

# 新建 2 個圖表
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 主要圖表
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # 子圖表

# 主圖表數據設定
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# 子圖表數據設定
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

# `fig, ax = plt.subplots( )` : 在一個畫布上繪製多個圖表
### plt.subplots( ) 等於結合以下2個動作：
**fig = plt.figure()：創建畫布**<br>
**ax = plt.subplot()：創建圖表**<br>

```python
x
```

```python
y
```

```python
fig, axes = plt.subplots()

# 可以直接在 axes 上建立數據
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

### `plt.subplots(nrows='行數', ncols='欄數')` : 可以指定行數和欄數

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # 1行,2欄
```

### 此時 axes 變成一個可以迭代的物件：

```python
axes
```

```python
axes[0].plot(x, y, 'r')
axes[0].set_title('First')

axes[1].plot(y, x, 'b')
axes[1].set_title('Second')
fig
```

### `plt.tight_layout()` : 加上這個會自動保持子圖之間的正確間距。

```python
# 重疊

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
```

# `.legend()`：在圖上標明一個圖例，用於說明每條曲線的文字顯示
### `圖表.plot(x, y, label)` 裡面要設置 label名稱，才會顯示在.legend( ) 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# 需要在.plot(x,y,label) 設置要顯示的 label
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

### `.legend(loc)` : 有一個 loc 函數，用來控制圖例位置
    loc = 0  : 'best' (最常用，讓程式決定最佳位置)
    loc = 1  : 'upper right'
    loc = 2  : 'upper left'
    loc = 3  : 'lower left'
    loc = 4  : 'lower right'
    loc = 5  : 'right'
    loc = 6  : 'center left'
    loc = 7  : 'center right'
    loc = 8  : 'lower center'
    loc = 9  : 'upper center'
    loc = 10 : 'center'

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location
fig
```

# 設置線顏色、線寬、線型 Setting colors, linewidths, linetypes

#### `線顏色`：
    'b' = 藍色 blue
    'r' = 紅色 red
    'g' = 綠色 green
    'y' = 黃色 yellow
    'k' = 黑色 black
    'w' = 白色 white
    'c' = 青色 cyan
    'm' = 品紅 magenta

#### `線種類`：
    '-'  = 實線 solid line style 
    '--' = 虛線 dashed line style
    '-.' = 點加線 dash-dot line style
    ':'  = 點線 dotted line style
    
#### `Markers種類`：
    '.' = 點 point marker
    ',' = 像素點 pixel marker
    'o' = 圓點 circle marker
    'v' = 向下的三角形 triangle_down marker
    '^' = 向上的三角形 triangle_up marker
    '<' = 向左的三角形 triangle_left marker
    '>' = 向右的三角形 triangle_right marker
    '1' = 向下叉點 tri_down marker
    '2' = 向上叉點 tri_up marker
    '3' = 向左叉點 tri_left marker
    '4' = 向右叉點 tri_right marker
    's' = 正方形 square marker
    'p' = 五角形 pentagon marker
    '*' = 星形 star marker
    'h' = 六邊形1 hexagon1 marker
    'H' = 六邊形2 hexagon2 marker
    '+' = 加號 plus marker
    'x' = X形狀 x marker
    'D' = 菱形 diamond marker
    'd' = 瘦菱形 thin_diamond marker
    '|' = 豎線 vline marker
    '_' = 橫線 hline marker

**可以運用類似 MATLAB 的寫法：**

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

**我們還可以通過名稱或 RGB 十六進制代碼定義顏色，並可以選擇使用 color 和 alpha 關鍵字參數提供 alpha 值:**<br>
`color` = 顏色<br>
`alpha` = 表示不透明度

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

## `.plot(數列1, 數列2, color, lw, ls, marker, markersize, markerfacecolor,`
## `markeredgewidth, markeredgecolor)`
#### `color` = '線顏色'
#### `lw` = 線寬度 linewidth
#### `ls` = 線類型 linestyle
#### `marker` = '標記'
#### `markersize` = 標記大小
#### `markerfacecolor` = '標記表面顏色'
#### `markeredgewidth` = 標記邊框寬
#### `markeredgecolor` = '標記邊框顏色'

### lw : 線寬度

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)
```

### ls : 線類型

```python
fig, ax = plt.subplots(figsize=(12,6))
# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
```

### 客製化點：

```python
fig, ax = plt.subplots(figsize=(12,6))

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...
```

### marker : 標記符號

```python
fig, ax = plt.subplots(figsize=(12,6))

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')
```

#### markersize : 標記大小 ,  markerfacecolor :  標記顏色 ,  markeredgewidth :  標記邊框寬度 ,   markeredgecolor : 標記邊框顏色

```python
fig, ax = plt.subplots(figsize=(12,6))

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

# 控制 x 軸、y 軸範圍
#### 方法一.    `set_xlim([xmin, xmax])`  /  `set_ylim([ymin, ymax])`
#### 方法二.    `.axis([xmin, xmax, ymin, ymax]) `
#### 方法三.    `.axis('tight') ` : 自動獲取緊密擬合的範圍

```python
fig, axes = plt.subplots(1, 4, figsize=(12, 4))



# 預設：不做任合設定
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")



# 用 .axis('tight') 自動獲取範圍
axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")



# 用 .axis([xmin, xmax, ymin, ymax]) 設定範圍
axes[2].plot(x, x**2, x, x**3)
axes[2].axis([2,5,0,60])
axes[2].set_title("custom axes range")



# 用 set_xlim([xmin, xmax]) / set_ylim([ymin, ymax]) 設定範圍
axes[3].plot(x, x**2, x, x**3)
axes[3].set_ylim([0, 60])
axes[3].set_xlim([2, 5])
axes[3].set_title("custom axes range");
```
